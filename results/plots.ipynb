```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.metrics import RocCurveDisplay, roc_auc_score

from zs_perturbation import DISEASES
```

# Plotting ROC curves

Here, we plot the ROC curve for each disease:


```python
roc_auc_scores = []

for disease in DISEASES:
    df = pd.read_csv(f"decoder/{disease}.csv")

    roc_auc_scores.append(roc_auc_score(df["y_true"], df["y_score"]))

    RocCurveDisplay.from_predictions(df["y_true"], df["y_score"], plot_chance_level=True, despine=True)
    plt.title(f"ROC Curve for {disease}")
```

# Average scores

This results in an average AUCROC score of 0.62:


```python
np.mean(roc_auc_scores)
```
